```python
def move_images_to_parent_folder(parent_folder):
  for species_folder in os.listdir(parent_folder):
    species_path = os.path.join(parent_folder, species_folder)
    if os.path.isdir(species_path):
      for image_file in os.listdir(species_path):
        image_path = os.path.join(species_path, image_file)
        if os.path.isfile(image_path):
          new_image_path = os.path.join(parent_folder, image_file)
          shutil.move(image_path, new_image_path)
      os.rmdir(species_path)


# Assuming your image folders are structured as described
edible_dir = "/content/mushroom-species/edible"
poisonous_dir = "/content/mushroom-species/poisonous"

move_images_to_parent_folder(edible_dir)
move_images_to_parent_folder(poisonous_dir)

print("Images moved successfully!")
```

```python
def split_data(source_dir, train_ratio=0.8, val_ratio=0.1, test_ratio=0.1):
  """Splits data into training, validation, and test sets.

  Args:
    source_dir: The directory containing the data.
    train_ratio: The proportion of data to use for training.
    val_ratio: The proportion of data to use for validation.
    test_ratio: The proportion of data to use for testing.
  """

  if train_ratio + val_ratio + test_ratio != 1.0:
    raise ValueError("Ratios must sum to 1.0")

  # Create target directories
  for split in ["train", "val", "test"]:
    for subdir in ["edible", "poisonous"]:
      os.makedirs(os.path.join(source_dir, split, subdir), exist_ok=True)

  for subdir in ["edible", "poisonous"]:
    source_subdir = os.path.join(source_dir, subdir)
    images = [f for f in os.listdir(source_subdir) if os.path.isfile(os.path.join(source_subdir, f))]
    random.shuffle(images)

    n_train = int(len(images) * train_ratio)
    n_val = int(len(images) * val_ratio)
    n_test = len(images) - n_train - n_val

    for i, image in enumerate(images):
      source_path = os.path.join(source_subdir, image)
      if i < n_train:
        destination_path = os.path.join(source_dir, "train", subdir, image)
      elif i < n_train + n_val:
        destination_path = os.path.join(source_dir, "val", subdir, image)
      else:
        destination_path = os.path.join(source_dir, "test", subdir, image)
      shutil.move(source_path, destination_path)


# Example usage
split_data("/content/mushroom-species")
```

```python
def count_images(directory):
  """Counts images in a directory and its subdirectories."""
  edible_count = 0
  poisonous_count = 0
  for root, _, files in os.walk(directory):
    for file in files:
      if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
        if "edible" in root:
          edible_count += 1
        elif "poisonous" in root:
          poisonous_count += 1
  return edible_count, poisonous_count


train_dir = "/content/mushroom-species/train"
val_dir = "/content/mushroom-species/val"
test_dir = "/content/mushroom-species/test"

# Count images in each split
train_edible, train_poisonous = count_images(train_dir)
val_edible, val_poisonous = count_images(val_dir)
test_edible, test_poisonous = count_images(test_dir)

# Print the results
print("Training Set:")
print(f"  Edible: {train_edible}")
print(f"  Poisonous: {train_poisonous}")
print(f"  Total: {train_edible + train_poisonous}")

print("\nValidation Set:")
print(f"  Edible: {val_edible}")
print(f"  Poisonous: {val_poisonous}")
print(f"  Total: {val_edible + val_poisonous}")

print("\nTest Set:")
print(f"  Edible: {test_edible}")
print(f"  Poisonous: {test_poisonous}")
print(f"  Total: {test_edible + test_poisonous}")
```

```python
train_dir = "/content/mushroom-species/train"
val_dir = "/content/mushroom-species/val"
test_dir = "/content/mushroom-species/test"
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet_v3.preprocess_input,
                                   rotation_range=20,
                                   horizontal_flip=True,
                                   fill_mode='nearest',
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   zoom_range=0.2,
                                   shear_range=0.2,
                                   brightness_range=[0.5, 1.5])

validation_datagen = ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet_v3.preprocess_input)
```

```python
# Menyiapkan data latih dan validasi dari kumpulan data gambar yang di-load dalam memori melalui fungsi flow()
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224,224),
    class_mode='categorical',
    shuffle=True,
    batch_size=16
)

val_generator = validation_datagen.flow_from_directory(
    val_dir,
    target_size=(224,224),
    class_mode='categorical',
    batch_size=128
)
```

```python
from tensorflow.keras.applications import MobileNetV3Large
from tensorflow.keras.models import Model

# Load MobileNetV2 pre-trained model
base_model = MobileNetV3Large(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = True
for layer in base_model.layers[:-50]:
    layer.trainable = False
```

```python
from tensorflow.keras import Sequential

model = Sequential([
    base_model,  # Add the pre-trained MobileNetV3 model
    tf.keras.layers.GlobalAveragePooling2D(),  # Pooling layer to flatten the output
    tf.keras.layers.Dense(256, activation='relu'),  # Fully connected layer
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(2, activation='softmax')  # Output layer for binary classification
])
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4), metrics=['accuracy'])
```

```python
lr_schedule = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss', factor=0.1, patience=3, min_lr=1e-6, verbose=1
)
```

```python
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, verbose=1, restore_best_weights=True)
```

```python
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10,
    callbacks=[lr_schedule, early_stop]
)
```

```python
# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python
from sklearn.metrics import classification_report

# Create a generator for the test data
test_generator = validation_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    class_mode='categorical',
    batch_size=1,  # Important: Set batch_size to 1 for prediction
    shuffle=False  # Important: Do not shuffle the data
)

# Predict the classes for the test data
predictions = model.predict(test_generator)

# Get the predicted classes
predicted_classes = np.argmax(predictions, axis=1)

# Get the true classes
true_classes = test_generator.classes

# Get the class labels
class_labels = list(test_generator.class_indices.keys())

# Print the classification report
print(classification_report(true_classes, predicted_classes, target_names=class_labels))
```
